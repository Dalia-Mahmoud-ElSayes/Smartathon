##Set up labels

```python
import csv
import shutil
def convert_to_labels(csv_file):
  annotation_file = open("/content/gdrive/MyDrive/Hackathon/"+csv_file+".csv")
  annotation_file = csv.reader(annotation_file, delimiter=" ")
  counter=0
  prev_img=''
  for annotation in annotation_file:
    labels=annotation[0].split(',')
    if counter<1:
        counter=counter+1
        continue    
    counter=counter+1
    # if(counter<11000):
    #    continue
    img_dim =[960,540]
    print(counter)
    bbox_x = (float(labels[4]) + float(labels[3])) / (2 * img_dim[0])  #x_min-x_max
    bbox_y = (float(labels[6]) + float(labels[5])) / (2 * img_dim[1])   #y_min-y_max
    bbox_w =(float(labels[3]) - float(labels[4])) / img_dim[0]    #x_max-x_min
    bbox_h = (float(labels[5]) - float(labels[6])) / img_dim[1]   #y_max-y_min
    class_id = labels[0]
    txt= str(class_id) +" "+ str(bbox_x) +" "+ str(bbox_y) +" "+ str(bbox_w) +" "+ str(bbox_h)

    fl_nm = labels[1]
    if(fl_nm==prev_img):
      fl_nm=fl_nm.replace("jpg","txt")
      with open("/content/gdrive/MyDrive/Hackathon/labels/"+csv_file+"/"+fl_nm,'a') as f:
        f.write('\n')
        f.write(txt)
      print("--")
    else:
      #move images to train folder
      #old="/content/gdrive/MyDrive/Hackathon/images/"+fl_nm
      #destination="/content/gdrive/MyDrive/Hackathon/images/"+csv_file+"/"+fl_nm
      #shutil.move(old,destination)
      prev_img=fl_nm
      fl_nm=fl_nm.replace("jpg","txt")
      with open("/content/gdrive/MyDrive/Hackathon/labels/"+csv_file+"/"+fl_nm,'w') as f:
        f.write(txt)
    
```

#Loading trained model

```python
import os
import torch
import pandas as pd
import csv
import numpy as np
#csv_result=pd.DataFrame()
imgs=[]
xmin=[]
ymin=[]
xmax=[]
ymax=[]
Class=[]
name=[]
image_path=[]
for img in os.listdir('/content/gdrive/MyDrive/Hackathon/images/test/'):
  text=str('/content/gdrive/MyDrive/Hackathon/images/test/'+img)
  imgs.append(text)
# print(imgs[0])


# Model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/gdrive/MyDrive/Hackathon/yolov5/runs/train/exp6/weights/best.pt', force_reload=True)

# Images
#imgs = ['/content/gdrive/MyDrive/Hackathon/images/test/006e04a34f54d116f6aff2bb31cb320c.jpg','/content/gdrive/MyDrive/Hackathon/images/test/0081b25823f23a536c9f1952491db636.jpg']  # batch of images
# type(imgs)
# Inference
for img in imgs:
  results = model(img)

# Results
  results.print()
  results.save()  # or .show()
# results.show()
  results.xyxy[0]  # img1 predictions (tensor)
  csv_result=results.pandas().xyxy[:]
  # print(type(csv_result[0]))

  if csv_result[0] is np.empty:
    continue
  else:
    count=0
    for row in csv_result[0]['xmin']:
      xmin.append(row)
      count=count+1
    for row in csv_result[0]['ymin']:
      ymin.append(row)
    for row in csv_result[0]['xmax']:
      xmax.append(row)
    for row in csv_result[0]['ymax']:
      ymax.append(row)
    for row in csv_result[0]['class']:
      Class.append(row)
    for row in csv_result[0]['name']:
      name.append(row)
    for i in range(count):
      image_path.append(img.replace("/content/gdrive/MyDrive/Hackathon/images/test/",""))
    

  
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(Class,image_path,name,xmax,xmin,ymax,ymin)),columns =[ 'class','image_path','name',  'xmax' , 'xmin' , 'ymax', 'ymin'])
df.to_csv('/content/gdrive/MyDrive/Hackathon/result.csv',index=False,header=True,encoding="UTF-8")
```
